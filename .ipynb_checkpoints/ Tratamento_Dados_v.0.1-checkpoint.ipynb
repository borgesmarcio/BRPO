```python
import pandas as pd
import xlrd

from datetime import datetime
import numpy as np
import seaborn as sns
```

```python

df = pd.read_excel("base_dados.xlsx", sheet_name = "2158151")
df_cheio =  pd.read_excel("base_dados.xlsx", sheet_name = None)

def trata_df(dataframe):
    df = dataframe.iloc[:, 0].str.split(';', expand=True)
    df.columns = [n.replace('"', '') for n in  dataframe.columns.str.split(';')[0]]
    return df

for i in range(len(list(df_cheio.keys()))-1):
    chaves = list(df_cheio.keys())
    df = pd.concat([df, df_cheio[chaves[i+1]]])

df_tratado = trata_df(df)
data = [xlrd.xldate_as_datetime(int(nume), 0).strftime("%d-%m-%Y") for nume in df_tratado["DATA_TRAT."]]
df_tratado['DATA_TRAT.'] = data
df_tratado["DATA_TRAT."] = [datetime.strptime(x, "%d-%m-%Y") for x in df_tratado["DATA_TRAT."]]
```

Transformando as colunas para valores numéricos para que possam ser agregadas

```python
def isfloat(str):
    try: 
        float(str)
    except ValueError: 
        return False
    return True
```

```python
# Retirando os dados que não tem preço pois não tiveram vendas 
df_tratado = df_tratado.loc[df_tratado["Preço"].str.replace(',', '.').apply(isfloat)]

# Tratando os dados  e fazendo casting para float
df_tratado["Vendas Reais Qtde"] = df_tratado["Vendas Reais Qtde"].astype("float")
df_tratado["Vendas Reais $"] = df_tratado["Vendas Reais $"].str.replace(",", ".").astype("float")
df_tratado["CMV"] = df_tratado["CMV"].str.replace(",", ".").astype("float")
df_tratado["preço2"] = df_tratado["preço2"].str.replace(",", ".").astype("float")
df_tratado["CMV/QTD"] = df_tratado["CMV/QTD"].str.replace(",", ".").astype("float")
df_tratado["Mg %"] = df_tratado["Mg %"].str.replace(",", ".").astype("float")
df_tratado["Margem Líquida Real %"] = df_tratado["Margem Líquida Real %"].str.replace(",", ".").astype("float")
df_tratado["Margem Líquida Real $"] = df_tratado["Margem Líquida Real $"].str.replace(",", ".").astype("float")
df_tratado["Venda Líquida Real $"] = df_tratado["Venda Líquida Real $"].str.replace(",", ".").astype("float")
df_tratado["Preço"] = df_tratado["Preço"].str.replace(",", ".").astype("float")

# Criando colunas de Semana e Ano para auxiliar na agregação
df_tratado.reset_index(inplace = True, drop = True)
df_tratado["Semana"] = [x.isocalendar()[1] for x in df_tratado["DATA_TRAT."]]
df_tratado["Ano"] = df_tratado["DATA_TRAT."].dt.year
```

Agrupando os dados por semana.

* Tira-se a média para as colunas CMV, preço2, CMV/QTD, MG%, Margem Líquida Real % e Preço
* Soma-se as demais outras colunas

```python
df_tratado
```

```python
df_agg = df_tratado[["SAP+NOME", "UF", "Ano", "Semana",  "preço2", "Mg %",
                     "Margem Líquida Real %", "Preço"]].groupby(["SAP+NOME", "UF", "Semana", "Ano"]).mean().reset_index()

df_agg_sum = df_tratado[["SAP+NOME", "UF", "Ano", "Semana", "Vendas Reais Qtde", "Vendas Reais $",
                     "Margem Líquida Real $", "Venda Líquida Real $", "CMV"]]. \
             groupby(["SAP+NOME", "UF", "Semana", "Ano"]).sum().reset_index()
df_agg_sum["CMV/QTD"] = df_agg_sum["CMV"] / df_agg_sum["Vendas Reais Qtde"]
```

Unindo os dois *dataframes* de agregação criados para chegar ao *dataset* que será depositado no *Data Lake*

```python
df_final = df_agg.copy()
df_final = df_final.merge(df_agg_sum, on = ["SAP+NOME", "UF", "Ano", "Semana"], how = "inner")
df_final
```

```python
df_final = df_final.loc[(df_final["Vendas Reais Qtde"] != 0.0) & (df_final["CMV/QTD"].notna())]
```

Gerando o *dataset* final:

```python
df_final.head()
```

```python
df_final.to_csv("df_rpo.csv", index = False)
```
