```python
import pandas as pd
from sdv.timeseries import PAR
```

```python
df = pd.read_csv("df_rpo_v2.csv")
```

```python
df = df[:1000]
```

```python
df["DATA_TRAT."] = pd.to_datetime(df["DATA_TRAT."], dayfirst = True)
```

```python
field_types = {"Preço": {
    "type": "numerical",
    "subtype": "float"
},
    "Vendas Reais Qtde": {
        "type": "numerical",
        "subtype": "integer"
    },
    "Vendas Reais $": {
        "type": 'numerical',
        "subtype": 'float'
    },
    "CMV": {
        "type": "numerical",
        "subtype": 'float'
    },
    "Margem Líquida Real %":{
        "type": "numerical",
        "subtype": "float"},
    "Margem Líquida Real $": {
        "type": "numerical",
        "subtype": "float"
    },
    "Preço": {
        "type": "numerical",
        "subtype": "float"
    },
    "preço2": {
        "type": "numerical",
        "subtype": "float"
    },
    "CMV/QTD": {  
        "type" : "numerical",
        "subtype": "float"
    },
    "MG": {
        "type" : "numerical",
        "subtype": "float"
    }
               
    }
```

```python
model = PAR(sequence_index = "DATA_TRAT.", 
           entity_columns = ["SAP+NOME", "UF"],
           sample_size = 10,
           field_types = field_types,
           epochs = 600,
           verbose = True)
```

```python
model.fit(df)
```

```python
novo_df = model.sample(5)
```
